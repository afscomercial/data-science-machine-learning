```python
#find the cust_category of bob marley
df2[df2['cust_name'] == 'Bob Marley']['cust_category']
```

```python
# find all the customers belonging to alberta
df2[df2['province'] == 'AB']
```

```python
# which customers belong to either Ontario or British Columbia
df2[(df2['province'] == 'ON') | (df2['province'] == 'BC')]
```

```python
# which customers belong to either Ontario or British Columbia using isin()
df2[df2['province'].isin(['ON', 'BC'])]
```

```python
# which are the distinct values of NumProducts in this dataset?
df2['num_products'].unique()
```

```python
# customers who have more than 2 products
df2[df2['num_products'] > 2]
```

```python
# which customers have numOfProducts, other than 2 or 3
df2[~df2['num_products'].isin([2, 3])]
```

```python
# customers whose names start with j
df2[df2['cust_name'].str.startswith('J')]
```

```python
# customers whose names have lowercase 'm'
df2[df2['cust_name'].str.contains('m')]
```

```python
# which customer has the longest relationship with the bank
df2[df2['len_relationship'] == df2['len_relationship'].max()]
```
